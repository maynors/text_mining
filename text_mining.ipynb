Treu les stopwords i realitza stemming al teu conjunt de dades.

En el ejercicio 1 nos hemos dado cuenta que hay muchos signos de puntuacin, palabras sin valor (stopwords) que crean ruido y no aportan al anlisis. En este ejercicio eliminaremos los signos y las palabras "stopwords".

### Palabras sin "," o "." por ejemplo

```python
onlywords_tokenized = [word for word in tokenized_word if word.isalpha()]
onlywords_dist = FreqDist(onlywords_tokenized)
onlywords_dist.tabulate(12)
```

> Ya hemos limpiado nuestro texto de signos de puntuacin, ahora falta eliminar las stopwords.

### Eliminar stopwords

```python
# Cargamos la librera y observamos lo que se consideran stopwords
from nltk.corpus import stopwords

stopWords = set(stopwords.words("english"))
print(stopWords)
```

```python
#Eliminamos las stopwords y mostramos lo resultante
words_token = []

for w in onlywords_tokenized:
    if w not in stopWords:
        words_token.append(w)

print(words_token)
```

```python
# Observamos cuantas palabras quedan despues sin stopwords
print("Palabras antes de eliminar stopwords",  (len(onlywords_tokenized)))
print("Palabras despues de eliminar stopwords",(len(words_token)))
```

> Se han reducido en un 46%.

### Reducir el texto mediante stemming

El stemming consiste en "unificar" aquellas palabras que son parecidas, por ejemplo: conocia, conocere, se puede sustituir a "conocer". En este apartado realizaremos este proceso y observaremos el resultado. Hay que tomar en cuenta el total de palabras no cambia, solo que hay menos palabras unicas.

```python
# Stemming
from nltk.stem import PorterStemmer

PStemmer = PorterStemmer()

text_stemmed = []
for x in words_token:
    text_stemmed.append(PStemmer.stem(x))

print(text_stemmed)
```

```python
# Confirmamos lo comentado anteriormente, el nmero de palabras es el mismo, lo que cambia son las palabras nicas.
print("Palabras antes de realizar stemming",  (len(words_token)))
print("Palabras despues de realizar stemming",(len(text_stemmed)))
```

### Visualizacin grfica despues de stemming y eliminar stopwords 

```python
#Graficamos las primeras 20 palabras ms repetidas

plt.title("Frequencia vs Palabras")
distStemmed = FreqDist(text_stemmed)
distStemmed.plot(20, cumulative=False)
```

# Exercici 3
Realitza sentiment analysis al teu conjunt de dades.

Para este ejercicio analizaremos el texto basado en el lxico.

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sentiment = SentimentIntensityAnalyzer()
sent_dictionary =sentiment.polarity_scores(text)

def sentiment_score(): 
    
    if sent_dictionary['compound'] > 0 :
        return 'Positive'
    elif sent_dictionary['compound'] <= - 0.011111 :
        return 'Negative'
    else :
        return 'Neutral'

result = sentiment_score()

print('El texto general es: ', sent_dictionary)
print('El texto fue calificado como', sent_dictionary['neg']*100, '% Negativo')
print('El texto fue calificado como', sent_dictionary['neu']*100, '% Neutro')
print('El texto fue calificado como', sent_dictionary['pos']*100, '% Positivo')

print('El texto en general fue calificado como',result, 'con una puntuacin de: ', sent_dictionary['compound'])
```

> El resultado muestra el porcentaje de palabras negativas, neutras y positivas. Al final, realiza un "compound", que suma y "normaliza" los valores entre -1 y 1, siendo 1 positivo y -1 negativo. En este caso, cataloga el texto como POSITIVO.

```python
#Recordar que en la variable tokenized_text almacena el texto dividido en oraciones
df = pd.DataFrame(tokenized_text, columns=['Oraciones'])
df.head(4)
```

```python
# Analizaremos por columna 
df = pd.DataFrame(tokenized_text, columns=['Sentence'])
df['Positive'] = [sentiment.polarity_scores(i)['pos'] for i in df['Sentence']]
df['Negative'] = [sentiment.polarity_scores(i)['neg'] for i in df['Sentence']]
df['Neutral'] = [sentiment.polarity_scores(i)['neu'] for i in df['Sentence']]
df['Compound'] = [sentiment.polarity_scores(i)['compound'] for i in df['Sentence']]
display(df)
```

```python
# Agregamos una columna

score = df['Compound'].values
sentimiento = []
for i in score:
    if i >= 0.05:
        sentimiento.append('Positive')
    elif i <= -0.05:
        sentimiento.append('Negative')
    else:
        sentimiento.append('Neutral')
        
df['sentimiento'] = sentimiento
df.head(10)
```

```python
import matplotlib.pyplot as plt

sent_counts = df['sentimiento'].value_counts().sort_values()
labels = sent_counts.index.tolist()
values = sent_counts.values.tolist()

# Crear el diagrama de pie
plt.title('Anlisis de oraciones en el texto')
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.show()
```

> Al analizar el texto en general, y tambin por oraciones. Se puede observar que el discurso esta lleno de positivismo. Sin embargo, al hacer enfsis en todas las desigualdades que han sufrido existen oraciones negativos.
